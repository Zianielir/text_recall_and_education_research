
#####**4.1 Отберите признаки, которые могут быть полезны при прогнозировании целевой (зависимой) переменной. Не включайте в число этих признаков переменную воздействия.**

```python
# Отбор данных
feat4_train = train.loc[:,data.columns.drop(['recall_score','higher_edu','far_from_uni'])]
feat4_test = test.loc[:,data.columns.drop(['recall_score','higher_edu','far_from_uni'])]
target4_train = train.loc[:,['recall_score']].values.ravel()
target4_test = test.loc[:,['recall_score']].values.ravel()
```

##### **4.2 Выберите произвольные значения гиперпараметров, а затем оцените и сравните(между методами) точность прогнозов с помощью RMSE и MAPE: на обучающей выборке.на тестовой выборке. с помощью кросс-валидации (используйте только обучающую выборку).**

```python
!pip install --upgrade scikit-learn
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# Определение моделей с произвольными гиперпараметрами
rf = RandomForestRegressor(random_state=52)
lr = LinearRegression()
knn = KNeighborsRegressor()

# Обучение моделей
rf.fit(feat4_train, target4_train)
lr.fit(feat4_train, target4_train)
knn.fit(feat4_train, target4_train)

# Предсказания на обучающей выборке
pred_rf_train = rf.predict(feat4_train)
pred_lr_train = lr.predict(feat4_train)
pred_knn_train = knn.predict(feat4_train)

# Предсказания на тестовой выборке
pred_rf_test = rf.predict(feat4_test)
pred_lr_test = lr.predict(feat4_test)
pred_knn_test = knn.predict(feat4_test)

# Оценка точности на обучающей выборке
mse_rf_train = mean_squared_error(target4_train, pred_rf_train)
rmse_rf_train = np.sqrt(mse_rf_train)
mse_lr_train = mean_squared_error(target4_train, pred_lr_train)
rmse_lr_train = np.sqrt(mse_lr_train)
mse_knn_train = mean_squared_error(target4_train, pred_knn_train)
rmse_knn_train = np.sqrt(mse_knn_train)

mape_rf_train = mean_absolute_percentage_error(target4_train, pred_rf_train)
mape_lr_train = mean_absolute_percentage_error(target4_train, pred_lr_train)
mape_knn_train = mean_absolute_percentage_error(target4_train, pred_knn_train)

# Оценка точности на тестовой выборке
mse_rf_test = mean_squared_error(target4_test, pred_rf_test)
rmse_rf_test = np.sqrt(mse_rf_test)
mse_lr_test = mean_squared_error(target4_test, pred_lr_test)
rmse_lr_test = np.sqrt(mse_lr_test)
mse_knn_test = mean_squared_error(target4_test, pred_knn_test)
rmse_knn_test = np.sqrt(mse_knn_test)

mape_rf_test = mean_absolute_percentage_error(target4_test, pred_rf_test)
mape_lr_test = mean_absolute_percentage_error(target4_test, pred_lr_test)
mape_knn_test = mean_absolute_percentage_error(target4_test, pred_knn_test)

# Кросс-валидация на обучающей выборке
cv_rf = cross_validate(rf, feat4_train, target4_train, cv=5, scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'])
cv_lr = cross_validate(lr, feat4_train, target4_train, cv=5, scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'])
cv_knn = cross_validate(knn, feat4_train, target4_train, cv=5, scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'])

# Вывод результатов
results = pd.DataFrame(
    {
        "Модель": ["Случайный лес", "Линейная регрессия", "K-ближайших соседей"],
        "RMSE (обучение)": [rmse_rf_train, rmse_lr_train, rmse_knn_train],
        "RMSE (тест)": [rmse_rf_test, rmse_lr_test, rmse_knn_test],
        "RMSE (кросс-валидация)": [
            -cv_rf["test_neg_root_mean_squared_error"].mean(),
            -cv_lr["test_neg_root_mean_squared_error"].mean(),
            -cv_knn["test_neg_root_mean_squared_error"].mean(),
        ],
        "MAPE (обучение)": [mape_rf_train, mape_lr_train, mape_knn_train],
        "MAPE (тест)": [mape_rf_test, mape_lr_test, mape_knn_test],
        "MAPE (кросс-валидация)": [
            -cv_rf["test_neg_mean_absolute_percentage_error"].mean(),
            -cv_lr["test_neg_mean_absolute_percentage_error"].mean(),
            -cv_knn["test_neg_mean_absolute_percentage_error"].mean(),
        ],
    }
)

results
```

##### **4.4 подберите на обучающей выборке оптимальные значения гиперпараметров градиентного бустинга ориентируясь на значение OOB (out-of-bag) ошибки. Сопоставьте гиперпараметры и точность на тестовой выборке для градиентного бустинга в зависимости от того, используется кросс-валидация или OOB ошибка**

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from itertools import product

# Задаем сетку гиперпараметров
param_grid_rf_oob = {
    "n_estimators": [100, 200, 300],
    "max_depth":    [2, 5, 8],
}

# Переменные для хранения лучшей модели
best_oob_rmse = float("inf")
best_oob_params = None
best_oob_model = None

# Ручной перебор по сетке и вычисление OOB RMSE
for n_estimators, max_depth in product(param_grid_rf_oob["n_estimators"],
                                       param_grid_rf_oob["max_depth"]):
    rf = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        oob_score=True,
        bootstrap=True,
        random_state=52,
        n_jobs=-1
    )
    rf.fit(feat4_train, target4_train)
    oob_pred = rf.oob_prediction_
    oob_rmse = np.sqrt(mean_squared_error(target4_train, oob_pred))

    print(f"OOB RMSE (n={n_estimators}, depth={max_depth}): {oob_rmse:.4f}")

    if oob_rmse < best_oob_rmse:
        best_oob_rmse = oob_rmse
        best_oob_params = {"n_estimators": n_estimators, "max_depth": max_depth}
        best_oob_model = rf

# Оценка на тестовой выборке
pred_oob_test = best_oob_model.predict(feat4_test)
rmse_oob_test = np.sqrt(mean_squared_error(target4_test, pred_oob_test))

# Сравнение с результатами GridSearchCV по CV (rf_grid уже есть)
# rf_grid.best_estimator_, rf_grid.best_score_ — из  GridSearchCV для CV

# Собираем результаты в таблицу
results = pd.DataFrame({
    "Модель": [
        "Случайный лес (кросс-валидация)",
        "Случайный лес (OOB-подбор)"
    ],
    "Гиперпараметры": [
        rf_grid.best_estimator_.get_params(),
        best_oob_model.get_params()
    ],
    "RMSE (обучение)": [
        -rf_grid.best_score_,            # отрицательный RMSE из GridSearchCV
        best_oob_rmse                    # OOB RMSE на train
    ],
    "RMSE (тест)": [
        np.sqrt(mean_squared_error(target4_test,
                                   rf_grid.best_estimator_.predict(feat4_test))),
        rmse_oob_test
    ]
})

print("\nСравнение подбора гиперпараметров:")
print(results.to_string())
```
